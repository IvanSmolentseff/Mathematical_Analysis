### **Задание 1:**

Рассмотрим последовательность  
$x_n =n^{(-1)^n}$.  

**а) Докажем, что последовательность неограниченная.**  
Пусть $n$ чётное: $x_n=n$, тогда $x_n$ положительна и увеличивается с ростом $n$.  
Пусть $n$ нечётное: $x_n=−n$, тогда $x_n$ отрицательна и убывает с ростом $n$.

Таким образом, последовательность $x_n$ неограниченная, так как модуль её членов стремится к бесконечности.

**б) Докажем, что последовательность не является бесконечно большой.**  
Поскольку последовательность $x_n$ чередуется между положительными и отрицательными значениями, она не является бесконечно большой, так как не существует такого числа $N$, что для всех $n>N$ выполняется условие $x_n>M$ или $x_n<-M$ для любого большого $M$.

### **Задание 2:**  
Докажем, что последовательность {sin $n$} расходится.

Последовательность {sin $n$} является неограниченной, так как значение синуса для различных значений $n$ (натуральных чисел) плотным образом заполняет отрезок $[−1,1]$. В частности, не существует конечного предела для этой последовательности, что означает её расходимость.

### **Задание 3:**  
Найти пределы:

а) $\lim\limits_{n\to\infty} \frac{10n}{n^2+1} = \lim\limits_{n\to\infty} \frac{10}{n+\frac{1}{n}} = 0$  

б)  $\lim\limits_{n\to\infty} \frac{n^2-n}{n-\sqrt{n}} = \lim\limits_{n\to\infty} \frac{n(n-1)}{n(1-\frac{1}{\sqrt{n}})} = \lim\limits_{n\to\infty} \frac{n-1}{1-\frac{1}{\sqrt{n}}} = \lim\limits_{n\to\infty} \frac{n-1}{1-0} =  \lim\limits_{n\to\infty} (n-1)= \infty $  
так как при $n→∞$, $\frac{1}{\sqrt{n}}$ стремится к нулю.  
в) $\lim\limits_{n\to\infty} \frac{5*3^n}{3^n-2} = \lim\limits_{n\to\infty} \frac{5}{1-\frac{2}{3^n}} = 5$

### **Задание 4:**
$\lim\limits_{n\to\infty} (\sqrt{n^2+n}-n) = \lim\limits_{n\to\infty} (\sqrt{n^2*(1+\frac{1}{n})}-n) = \lim\limits_{n\to\infty} (n*\sqrt{1+\frac{1}{n}}-n) = \lim\limits_{n\to\infty} n(\sqrt{1+\frac{1}{n}}-1) $

Теперь упростим выражение под корнем с помощью разложения в ряд Тейлора для
$\sqrt{1+x}$ при малых $x$:  
$\sqrt{1+\frac{1}{n}} ≈ 1 + \frac{1}{2n} $  
Подставим это обратно:  
$\lim\limits_{n\to\infty} n(1+\frac{1}{2n}-1) = \lim\limits_{n\to\infty} (\frac{n}{2n}) = \frac{1}{2} $

***Таким образом, предел:***  
$\lim\limits_{n\to\infty} (\sqrt{n^2+n}-n) = \frac{1}{2} $
