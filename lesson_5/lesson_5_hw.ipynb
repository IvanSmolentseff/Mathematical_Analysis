## 1. Найти пределы


**1. $\lim\limits_{x\to\infty} [\ln(x+3) - \ln{x}]$**   

Применим свойства логарифмов:  
$\ln(x+3)−\ln{x}=\ln(\frac{x+3}{x}​)=\ln(1+\frac{3}{x}​)$  
Так как при $x→∞$ $\frac{3}{x}→0$, то:  
$\lim\limits_{xn\to\infty}\ln(1+\frac{3}{x}) = \ln(1) = 0$  

**Ответ:**  
$\lim\limits_{x\to\infty} [\ln(x+3) - \ln{x}] = 0$



**2. $\lim\limits_{x\to 0} \frac{\ln(1+2x)}{\arcsin{3x}}$**  

Используем приближенные разложения для малых $x$:  
$\ln(1+2x)≈2x$ и $\arcsin⁡(3x)≈3x$, когда $x→0$. Тогда:  
$\frac{\ln(1+2x)}{\arcsin{3x}} ≈ \frac{2x}{3x} = \frac{2}{3}$  

**Ответ:**  
$\lim\limits_{x\to 0} \frac{\ln(1+2x)}{\arcsin{3x}} = \frac{2}{3}$


**3. $\lim\limits_{x\to 0} \frac{7^x-1}{3^x-1}$**  

Для решения этого предела можно воспользоваться разложением экспоненциальных функций при малых $x$. Известно, что $ax≈1+x\ln{a}$ при $x→0$. Применим это разложение:  
$7^x≈1+x\ln(7)$,   $3^x≈1+x\ln(3)$  
Подставим эти приближенные выражения в исходный предел:  
$\frac{7^x-1}{3^x-1} ≈ \frac{1+x\ln(7)-1}{1+x\ln(3)-1} = \frac{x\ln(7)}{x\ln(3)} = \frac{\ln(7)}{\ln(3)} $  

**Ответ:**  
$\lim\limits_{x\to 0} \frac{7^x-1}{3^x-1}= \frac{\ln(7)}{\ln(3)} $


**4. $\lim\limits_{a\to 0} \frac{(x+a)^3-x^3}{a}$**    

Это производная функции $f(x)=x^3$ по $x$ по определению производной:  
$f'(x)=\lim\limits_{a\to 0} \frac{(x+a)^3-x^3}{a}=3x^2$  

**Ответ:**  
$\lim\limits_{a\to 0} \frac{(x+a)^3-x^3}{a}=3x^2$

**5. $\lim\limits_{x\to\infty} (\frac{x^3}{5x^2+1} - \frac{x^2}{5x-3})$**  

Приведем выражения к более простому виду:  
$\frac{x^3}{5x^2+1} ≈ \frac{x^3}{5x^2} = \frac{x}{5}$  
$\frac{x^2}{5x-3} ≈ \frac{x^2}{5x}= \frac{x}{5}$  

Тогда:  
$\lim\limits_{x\to\infty} (\frac{x}{5} - \frac{x}{5}) = 0$  

**Ответ:**  
$\lim\limits_{x\to\infty} (\frac{x^3}{5x^2+1} - \frac{x^2}{5x-3})=0$

**6. $\lim\limits_{x\to 0} \frac{1-\cos{4x}}{2x\tan{2x}}$**

Начнем с использования стандартных приближений для малых $x$. Для косинуса и тангенса можно использовать следующие приближения:  
$1-\cos{4x} ≈ \frac{(4x)^2}{2} = 8x^2$  
$\tan{2x} ≈ 2x$  
Теперь подставим эти выражения в предел:  
$\frac{1-\cos{4x}}{2x\tan{2x}} ≈ \frac{8x^2}{2x*2x} = 2$

**Ответ:**  
$\lim\limits_{x\to 0} \frac{1-\cos{4x}}{2x\tan{2x}} = 2$

**7. $\lim\limits_{x\to\infty} x\sin(\frac{2}{x})$**   

Для решения этого предела воспользуемся приближением $\sin{y}≈y$ при малых $y$. Так как $\frac{2}{x}→0$ при $x→∞$, то:  
$\sin(\frac{2}{x}​)≈\frac{2}{x}$  
Теперь подставим это приближение в исходный предел:  
$x\sin(\frac{2}{x}​) ≈ x⋅\frac{2}{x} = 2$  

**Ответ:**  
$\lim\limits_{x\to\infty} x\sin(\frac{2}{x})=2$

**8. $\lim\limits_{x\to 0} (1+\tan{x})^{\cot{x}}$**  

При $x→0$ используется следующее приближение: $\tan{x}≈x$ и $\cot{x}=\frac{1}{x}$.  
Запишем выражение с этими приближениями:  
$(1+\tan{x})^{\cot{x}} ≈ (1+x)^{\frac{1}{x}}$  
Этот предел является классическим, и известно, что:  
$\lim\limits_{x\to 0}(1+x)^{\frac{1}{x}} = e$

**Ответ:**  
$\lim\limits_{x\to 0} (1+\tan{x})^{\cot{x}}=e$

**9. $\lim\limits_{x\to 0} (\cos{2x})^{\frac{1}{\sin^2{x}}}$**  

Начнем с приближения $\cos{2x}≈1-2x^2$ при $x→0$. Для малых $x$, $\sin{x}≈x$, так что $\sin^2{x}≈x^2$.  
Теперь преобразуем выражение:  
$(\cos{2x})^{\frac{1}{\sin^2{x}}} ≈ (1-2x^2)^{\frac{1}{x^2}}$  
Здесь возникает стандартный предел вида $(1-u)^{1/u}$ при $u→0$, который равен $e^{-2}$. В нашем случае $u=2x^2$, и при $x→0$ этот предел стремится к $e^{-2}$.  

Ответ:  
$\lim\limits_{x\to 0} (\cos{2x})^{\frac{1}{\sin^2{x}}} = e^{-2}$

**10. $\lim\limits_{x\to 0} \frac{\sqrt{1+x\sin{x}}-1}{x^2}$**  

Для малых $x$ используем приближение $\sin{x}≈x$, тогда:  
$1+x\sin{x}≈1+x^2$  
Теперь применим разложение для корня: $\sqrt{1=y}≈1+\frac{y}{2}$ при малых $y$. Тогда:   
$\frac{\sqrt{1+x\sin{x}}-1}{x^2} ≈ \frac{(1+\frac{x^2}{2})-1}{x^2} = \frac{\frac{x^2}{2}}{x^2} = \frac{1}{2}$  

**Ответ:**  
$\lim\limits_{x\to 0} \frac{\sqrt{1+x\sin{x}}-1}{x^2}= \frac{1}{2}$

## 2. Установить характер разрыва функции в точке $x₀$:


**1. $f(x) = \frac{x^2-16}{x+4}$, $x_0=-4$**   
Для начала попробуем упростить функцию $f(x)$.   
$f(x) = \frac{x^2-16}{x+4} = \frac{(x-4)(x+4)}{x+4}$  
Для $x≠-4$ можно сократить $x+4$, получим:   
$f(x) = x-4$  
Но при $x=-4$ знаменатель обращается в ноль, следовательно, функция имеет точку разрыва в $x=-4$. Рассмотрим пределы слева и справа.  
$\lim\limits_{x\to -4} f(x) = (-4) - 4 = -8$, $\lim\limits_{x\to -4} f(x) = (-4) - 4 = -8$  
Пределы слева и справа совпадают, но функция не определена в $x=-4$. Cледовательно, разрыв устранимый.

**Ответ:**  
разрыв устранимый в $x_0=-4$.

**2. $f(x)=\frac{sin{x}}{x}$, $x_0=0$**  

Здесь мы имеем классическую неопределенность вида $\frac{0}{0}$. Рассмотрим предел функции при $x→0$:  
$\lim\limits_{x\to 0}\frac{sin{x}}{x} = 1$  
Следовательно, предел существует и равен 1, однако $f(x)$ не определена в точке $x=0$. Разрыв устранимый, так как если определить $f(0)=1$, функция станет непрерывной в этой точке.  

**Ответ:**  
разрыв устранимый в $x_0=0$.

## 3. Исследовать на непрерывность функцию $f(x)$ в точке $x_0$:


**1. $f(x)=\arctan{\frac{2}{x-1}}$, $x_0=1$**  

Для исследования непрерывности функции в точке $x_0=1$, нужно сначала проверить, существует ли предел функции при $x→1$, а также значение функции в этой точке.

*Шаг 1: Определим значение функции в точке $x_0=1$:*  
$f(1)= \arctan{(\frac{2}{1-1})}$  
Но здесь возникает деление на ноль, значит функция не определена в точке $x=1$.  

*Шаг 2: Исследуем предел функции при $x→1$:*  
* При $x→1^-$ (слева от 1): $\frac{2}{x-1}→−∞$, поэтому:  
$\lim\limits_{x\to 1^-} f(x) = \arctan{−∞} = -\frac{2}{\pi}​$  
* При $x→1^+$ (справа от 1): $\frac{2}{x-1}→+∞$, поэтому:  
$\lim\limits_{x\to 1^+} f(x) = \arctan{+∞} = \frac{2}{\pi}​$    

Вывод: Пределы слева и справа от точки $x_0=1$ не совпадают, следовательно, функция имеет разрыв второго рода в этой точке.  

**Ответ:**  
разрыв второго рода в $x_0=1$.

**2. $f(x) = \frac{1}{2^{x-3}-1}$, $x_0=3$**  

*Шаг 1: Найдем значение функции в точке $x_0=3$*  
$f(3) = \frac{1}{2^{3-3}-1} = \frac{1}{2^0-1} = \frac{1}{1-1}$  
Здесь возникает деление на ноль, что указывает на то, что функция не определена в точке $x=3$. Это означает, что точка $x_0=3$ — это потенциальная точка разрыва.  

*Шаг 2: Исследуем предел функции при $x→3$*  
* При $x→3^-$ (слева от 3): $2^{x-3}→1$, но $2^{x-3}-1→0^-$ (меньше нуля), поэтому  
$\lim\limits_{x\to 3^-} f(x) = \lim\limits_{x\to 3^-} \frac{1}{2^{x-3}-1} →−∞$  
* При $x→3^+$ (справа от 3): $2^{x-3}→1$, но $2^{x-3}-1→0^+$ (больше нуля), поэтому  
$\lim\limits_{x\to 3^-} f(x) = \lim\limits_{x\to 3^-} \frac{1}{2^{x-3}-1} →+∞$

Вывод: пределы слева и справа от точки $x_0=3$ расходятся (один уходит в $-∞$, а другой в $+∞$). Это означает, что в точке $x_0=3$ имеет место разрыв второго рода.  

**Ответ:**  
разрыв второго рода в $x_0=3$.
