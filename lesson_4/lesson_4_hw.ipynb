## Тема “Предел функции”

### 1. Предложить пример функции, не имеющей предела в нуле и в бесконечностях.

$f(x)=sin(\frac{1}{x}​)$

**Обоснование:**  

**При $x→0$:**   
Когда $x$ приближается к нулю, значение $\frac{1}{x}$ стремится к бесконечности. Поскольку синус — периодическая функция, то $sin(\frac{1}{x}​)$  будет осциллировать между значениями от -1 до 1, и, соответственно, не будет иметь предела.

**При $x→∞$ и $x→−∞$:**   
В этих случаях $\frac{1}{x}$ стремится к нулю, и $sin(\frac{1}{x}​)$ стремится к $sin(0)=0$. Однако это не изменяет того факта, что в точке $x=0$ функция все равно не имеет предела.


Таким образом, функция $sin(\frac{1}{x}​)$ не имеет предела ни при $x→0$, ни при $x→∞$ и $x→−∞$

### 2. Привести пример функции, не имеющей предела в точке, но определенной в ней.

$f(x)=sin(\frac{1}{x}​)$ при $x\neq0$ и $f(0)=0$

**Обоснование:**  

**Определённость в точке $x=0$:**  
Функция $f(x)$ определена в точке $x=0$ таким образом: $f(0)=0$.

**Отсутствие предела в точке $x=0$:**  
При приближении $x$ к нулю (с любой стороны), значение $\frac{1}{x}$ начинает стремительно увеличиваться по абсолютной величине, и функция $sin(\frac{1}{x}​)$ начинает бесконечно осциллировать между -1 и 1. Из-за этих осцилляций функция
$f(x)$ не имеет определённого предела в точке $x=0$, хотя она и остаётся определенной в этой точке.


### 3. Исследовать функцию $f(x) = x^3 - x^2$ по плану:  
a. Область задания и область значений.  
b. Нули функции и их кратность.  
c. Отрезки знакопостоянства.  
d. Интервалы монотонности.  
e. Четность функции.  
f. Ограниченность.  
g. Периодичность.

**a. Область задания и область значений.**

**Область задания:**   
Функция $f(x) = x^3 - x^2$ определена на всей числовой прямой, так как она является полиномом. Следовательно, область задания $D(f)=\mathbb{R}$ (все действительные числа).

**Область значений:**   
Так как функция представляет собой кубический полином, её область значений также $\mathbb{R}$ (все действительные числа). В этом случае, функция может принимать любые значения, так как $x^3$ доминирует при больших значениях $∣x∣$.

**b. Нули функции и их кратность.**  

Нули функции находятся из уравнения $f(x)=0$:  
$x^3−x^2=x^2(x−1)=0$  
Решая это уравнение, получаем:  
$x^2=0$ или $x−1=0$  
**Нули функции:**  
$x=0$ (с кратностью 2, так как $x^2=0)$.  
$x=1$ (с кратностью 1).

**c. Отрезки знакопостоянства.**  

Исследуем знаки функции на промежутках, определяемых нулями $x=0$ и $x=1$:  
1. $x<0:$  
$f(x) = x^2(x - 1)$  
Для $x<0$, $x^2>0$ и $x−1<0$, следовательно, $f(x)<0$.  

2. $0<x<1:$  
$x^2>0$ и $x−1<0$, следовательно, $f(x)<0$.  

3. $x>1:$  
$x^2>0$ и $x−1>0$, следовательно, $f(x)>0$.

*Итак, функция отрицательна на интервалах $(−∞,0)$ и $(0,1)$ и положительна на интервале $(1,∞)$.*


**d. Интервалы монотонности.**  

Для нахождения интервалов монотонности рассмотрим производную функции:  
$f'(x) = 3x^2 - 2x$  
Найдем критические точки:  
$f'(x) = x(3x - 2)$   
Критические точки:  
$x=0$ и $x=\frac{2}{3}$  

Теперь исследуем знак производной на интервалах:  
1. $x<0:$  
$f'(x)<0$, значит функция убывает.

2. $0<x<\frac{2}{3}:$  
$f'(x)<0$, функция убывает.  

3. $x>\frac{2}{3}:$  
$f'(x)>0$, функция возрастает.  

*Итак, функция убывает на интервалах $(−∞,\frac{2}{3})$ и возрастает на интервале $(\frac{2}{3},∞)$.*

**e. Четность функции.**  

Функция $f(x)=x^3-x^2$ не является четной, поскольку $f(-x)=-x^3-x^2 \neq f(x)$. Она также не является нечетной, так как $f(-x)\neq -f(x)$

**f. Ограниченность.**  

Функция не является ограниченной сверху или снизу, так как при больших положительных или отрицательных значениях $x$, значение функции также стремится к бесконечности или минус бесконечности соответственно.

**g. Периодичность.**  

Функция $f(x) = x^3 - x^2$ не является периодической, так как это полином третьей степени, и его график не повторяется с какой-либо периодичностью.

### **a.** Чтобы найти предел $\lim\limits_{n\to 0} \frac{3x^3 - 2x^2}{4x^2}$, упростим выражение.  
$\lim\limits_{n\to 0} \frac{3x^3 - 2x^2}{4x^2} = \lim\limits_{n\to 0} \frac{x^2(3x-2)}{4x^2} = \lim\limits_{n\to 0} \frac{3x-2}{4}$  

Подставляем $x=0$ в оставшееся выражение:  
$\frac{3(0)-2}{4} = \frac{-2}{4} = -\frac{1}{2}$  

Таким образом, предел равен:
$\lim\limits_{n\to 0} \frac{3x^3 - 2x^2}{4x^2} = -\frac{1}{2}$

### **b.** Для нахождения предела $\lim\limits_{n\to 0} \frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1}$, сначала рассмотрим числитель и знаменатель отдельно.

*Числитель* $\sqrt{1+x}-1$   
При $x→0$ это выражение имеет неопределённость вида $\frac{0}{0}$, поэтому можно применить разложение в ряд Тейлора или линеаризацию:  
$\sqrt{1+x}≈1+\frac{x}{2}$  
Таким образом, числитель:  
$\sqrt{1+x}-1≈\frac{x}{2}$  


*Знаменатель* $\sqrt[3]{1+x}-1$  
Аналогично, при $x→0$ это выражение имеет неопределённость вида $\frac{0}{0}$. Разложение в ряд Тейлора или линеаризация для кубического корня:  
$\sqrt[3]{1+x}-1 ≈\frac{x}{3} $  


*Теперь подставим эти приближённые значения в предел*:  
$\lim\limits_{n\to 0} \frac{\frac{x}{2}}{\frac{x}{3}} = \lim\limits_{n\to 0} \frac{3x}{2x} = \frac{3}{2}$
  

*Таким образом, предел равен:*  
$\lim\limits_{n\to 0} \frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1} = \frac{3}{2}$

### **c.** Для нахождения предела $\lim\limits_{n\to\infty} (\frac{x+3}{x})^{4x+1}$, упростим выражение внутри скобок и потом найдем его предел.  

$\frac{x+3}{x} = 1 +\frac{3}{x}$  
Таким образом, предел можно переписать как:  
$\lim\limits_{n\to\infty} (1 +\frac{3}{x})^{4x+1}$  

*Переход к экспоненциальной форме*  
Заметим, что данное выражение похоже на предел для экспоненты $e$, а именно:  
$\lim\limits_{n\to\infty} (1+\frac{a}{n})^n = e^a$  

*Для использования этой формулы, перепишем наш предел:*  
$\lim\limits_{n\to\infty} [(1 +\frac{3}{x})^x]^{4+\frac{1}{x}}$  

Так как $(1 +\frac{3}{x})^x$  при $x→∞$ стремится к $e^3$, наш предел становится:  
$(e^3)^4 = e^{12}$  

*Следовательно, предел равен:*  
$\lim\limits_{n\to\infty} (\frac{x+3}{x})^{4x+1} = e^{12}$

$\lim\limits_{n\to 0} \frac{\sin(x)}{x} = 1$  
Инвертируя этот предел, получаем:  
$\lim\limits_{n\to 0} \frac{x}{\sin(x)} = \frac{1}{\lim\limits_{n\to 0} \frac{\sin(x)}{x}} = \frac{1}{1} = 1$

Для нахождения предела $\lim\limits_{n\to 0} \frac{x}{\sin(x)}$ , можно воспользоваться известным пределом:

Инвертируя этот предел, получаем:  
$\lim\limits_{n\to 0} \frac{x}{\arcsin(x)} = \frac{1}{\lim\limits_{n\to 0} \frac{\arcsin(x)}{x}} = \frac{1}{1} = 1$

$\lim\limits_{n\to\infty} (\frac{4x+3}{4x-3})^{6x} = \lim\limits_{n\to\infty} (\frac{4x(1+\frac{3}{4x})}{4x(1-\frac{3}{4x})})^{6x} = \lim\limits_{n\to\infty} (\frac{1+\frac{3}{4x}}{1-\frac{3}{4x}})^{6x}$  

рассмотрим предел для $\frac{1+\frac{3}{4x}}{1-\frac{3}{4x}}$ при $x→∞:$

Таким образом, наш предел принимает вид:  
$\lim\limits_{n\to\infty} (1 + \frac{6}{4x})^{6x}$

$\lim\limits_{n\to\infty} (1 + \frac{a}{x})^{bx} = e^{ab}$  

$e^{6*\frac{6}{4}} = e^9$  

Таким образом, ответ на предел:
$\lim\limits_{n\to\infty} (\frac{4x+3}{4x-3})^{6x} = e^9$

$\lim\limits_{n\to\infty} \frac{\sin{x} + \ln{x}}{x} = \lim\limits_{n\to\infty} (\frac{\sin{x}}{x} + \frac{\ln{x}}{x})$
