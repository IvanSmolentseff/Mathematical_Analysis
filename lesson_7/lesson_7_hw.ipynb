```python
import numpy as np
import matplotlib.pyplot as plt

# 1) y = ln(1 - x^2)
x1 = np.linspace(-0.99, 0.99, 500)
y1 = np.log(1 - x1**2)

# 2) y = x^2 / (1 - x^2)
x2 = np.linspace(-0.99, 0.99, 500)
y2 = x2**2 / (1 - x2**2)

# 3) y = x^2 * e^(-x)
x3 = np.linspace(0, 5, 500)
y3 = x3**2 * np.exp(-x3)

# 4) y = x - ln(x)
x4 = np.linspace(0.01, 5, 500)
y4 = x4 - np.log(x4)

# Create the plots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot for y = ln(1 - x^2)
axs[0, 0].plot(x1, y1, label=r"$y = \ln(1 - x^2)$")
axs[0, 0].set_title(r"$y = \ln(1 - x^2)$")
axs[0, 0].axvline(x=-1, color='r', linestyle='--')
axs[0, 0].axvline(x=1, color='r', linestyle='--')
axs[0, 0].grid(True)

# Plot for y = x^2 / (1 - x^2)
axs[0, 1].plot(x2, y2, label=r"$y = \frac{x^2}{1 - x^2}$")
axs[0, 1].set_title(r"$y = \frac{x^2}{1 - x^2}$")
axs[0, 1].axvline(x=-1, color='r', linestyle='--')
axs[0, 1].axvline(x=1, color='r', linestyle='--')
axs[0, 1].grid(True)

# Plot for y = x^2 * e^(-x)
axs[1, 0].plot(x3, y3, label=r"$y = x^2 e^{-x}$")
axs[1, 0].set_title(r"$y = x^2 e^{-x}$")
axs[1, 0].grid(True)

# Plot for y = x - ln(x)
axs[1, 1].plot(x4, y4, label=r"$y = x - \ln(x)$")
axs[1, 1].set_title(r"$y = x - \ln(x)$")
axs[1, 1].grid(True)

# Show plots
plt.tight_layout()
plt.show()
```
